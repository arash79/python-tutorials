```python
'''matplotlib is one the most famous libraries for plotting and visualizing data in python. it has different types of plotting
and one can choose from them arbitrarily.'''
'''first let's simply plot some points.'''
```

```python
import matplotlib.pyplot as plt
x = [10, 20, 30, 40]
y = [20, 30, 40, 50]
plt.plot(x, y)
plt.show()
```

```python
'''now let's complicate things a little bit 😄. what if i want to add a title to my plot?'''
x = [10, 20, 30, 40]
y = [20, 30, 40, 50]
plt.plot(x, y)
plt.title("this is my title.")
plt.show()
```

```python
'''and let's make it even more complicated 🤪. what if i want to add a label to either of my axes? or even one of them?'''
x = [10, 20, 30, 40]
y = [20, 30, 40, 50]
plt.plot(x, y)
plt.xlabel("this is my X axis.")
plt.ylabel("this is my Y axis.")
plt.title("and this is the title we've talked about.")
plt.show()
```

```python
'''and what if we want to plot a specific region of a chart, how we do it? this is how we do it.'''
x = [1, 4, 5, 12]
y = [2, 5, 7, 20]
plt.plot(x, y)
plt.axis([0, 15, 0, 25])
'''Note that the first two elements of the axis method argument is the X interval and the later two are Y interval.'''
plt.show()
```

```python
'''we can set X axis step tickers arbitrarily using xticks() method. we can do the same with the Y axis too.'''
x = [1, 4, 5, 12]
y = [2, 5, 7, 20]
plt.plot(x, y)
plt.axis([0, 15, 0, 25])
plt.xticks([i for i in range(15)])
plt.yticks([i for i in range(0, 26, 3)])
plt.show()
```

```python
'''lets create a function example with everything we've learned.'''
x = [i for i in range(10)]
y = [pow(i, 2) for i in x]
plt.plot(x, y)
plt.title("this is a quadratic parabola.")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.xticks([i for i in range(10)])
plt.yticks([i for i in range(0, 100, 5)])
plt.show()
```

```python
'''but how can we create a subplot from an originial plot. take sine function as an example and look at below exercise. here
we will need to import numpy too.'''
import numpy as np
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)
'''in our elementary example we would plot this as:'''
plt.plot(x, y)
plt.show()
```

```python
'''but we said that we want to creat subplots. it's a little trickier. pay attention to the syntax differences.'''
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)
figure, axis = plt.subplots()
axis.plot(x, y)
'''we add title to the plots by set_title() in this method of plotting. so by now we have set_title() VS title()'''
axis.set_title('main plot.')
plt.show(figure)
```

```python
'''the above example resulted in the main plot itself with no difference with the plt.plot() method. we want to create subplots.
this is the way:'''
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)
figure, axes = plt.subplots(2)
'''Note how we set title for the plot in this method. suptitle()'''
figure.suptitle("Vertically stacked subplots")
axes[0].plot(x, y)
axes[1].plot(x, -y)
axes[0].set_title("Title")
plt.show(figure)
```

```python
'''If you are creating just a few Axes, it's handy to unpack them immediately to dedicated variables for each Axes.'''
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)
figure, (axis_1, axis_2) = plt.subplots(2)
figure.suptitle("Vertically stacked subplots")
axis_1.plot(x, y)
axis_2.plot(x, -y)
plt.show(figure)
```

```python
'''we can do same with other orientations.'''
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)
figure, (axis_1, axis_2) = plt.subplots(1, 2)
figure.suptitle('Horizontally stacked subplots')
axis_1.plot(x, y)
axis_2.plot(x, -y)
plt.show(figure)
```

```python
'''we can also have mutiple subplots stacking on each other in either directions.'''
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)
figure, axes = plt.subplots(2, 2)
figure.suptitle('2-D stacked subplots')
axes[0, 0].plot(x, y)
'''notice here that how we can set each subplot title individually'''
axes[0, 0].set_title('first subplot')
axes[0, 1].plot(x, y, 'tab:orange')
axes[0, 1].set_title('second subplot')
axes[1, 0].plot(x, -y, 'tab:green')
axes[1, 0].set_title('third subplot')
axes[1, 1].plot(x, -y, 'tab:red')
axes[1, 1].set_title('fourth subplot')
axes[1, 1].set_xlim([2, 4])
axes[1, 1].set_ylim([0, 1])
'''here we add X and Y axis labels collectively.'''
for axis in axes.flat:
    axis.set(xlabel='x-label', ylabel='y-label')
'''but we can set X and Y labels individally as:'''
axes[0, 1].set_xlabel("arbitrary X label")
axes[0, 1].set_ylabel("arbitrary Y label")
plt.show(figure)
```

```python
'''and the last important thing that we need to talk about is how to restrict the X and Y axis limits and get a zoomed view.'''
'''here is the simple plot example.'''
x = np.linspace(0, 100, 1000)
y = np.power(x, 0.5)
plt.plot(x, y)
plt.title("this is the whole view.")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```

```python
'''now we want to limit the X and Y axis in the above example and get a zoomed view in an specific region.'''
x = np.linspace(0, 100, 1000)
y = np.power(x, 0.5)
plt.plot(x, y)
plt.title("this is a zoom-in view.")
plt.xlabel("X axis limited between 2 and 4")
plt.ylabel("Y axis limited between 0 and 5")
plt.xlim([2, 4])
plt.ylim([0, 5])
plt.show()
'''the same would be applicable to subplots but have in mind that almost every function that we use in simple plots are
available in subplots with a slight difference in syntax and that is: in subplots there comes a "set" in the beginning of the
method. for example for the xlim() we have set_xlim()'''
```

```python
'''now let's learn scatter plotting. have in mind everything we've learned so far is also applicable in this type of plotting'''
x = [5, 2, 9, 4, 7] 
y = [10, 5, 8, 4, 2]  
plt.scatter(x, y) 
plt.title("this is a scatter plot")
plt.xlabel("this is the X axis")
plt.ylabel("this is the Y axis")
plt.show()
```

```python
'''one can change the marker shape in scatter plot. you can see different types of markers in: 
https://matplotlib.org/stable/api/markers_api.html#module-matplotlib.markers'''
x = [5, 2, 9, 4, 7] 
y = [10, 5, 8, 4, 2]  
plt.scatter(x, y, marker="*") 
plt.title("this is a scatter plot")
plt.xlabel("this is the X axis")
plt.ylabel("this is the Y axis")
plt.show()
```

```python
'''you can also change the marker color. notice that color is not exclusive to scatter plot and it can be used in every type of 
plot that you desire.'''
x = [5, 2, 9, 4, 7] 
y = [10, 5, 8, 4, 2]  
plt.scatter(x, y, marker="+", c="r") 
plt.title("this is a scatter plot")
plt.xlabel("this is the X axis")
plt.ylabel("this is the Y axis")
plt.show()
```

```python
"and here is how we draw bar charts in matplotlib."
x = [5, 2, 9, 4, 7] 
y = [10, 5, 8, 4, 2]  
plt.bar(x, y) 
plt.title("this is a bar chart")
plt.xlabel("this is the X axis")
plt.ylabel("this is the Y axis")
plt.show()
```

```python
'''we can also have multiple plots in a single figure.'''
x = np.linspace(0, 100, 1000)
y = np.power(x, 0.5)
x_ = [5, 2, 9, 4, 7] 
y_ = [10, 5, 8, 4, 2]  
plt.plot(x, y, c="g")
plt.scatter(x_, y_, c="y")
plt.show()
```

```python
'''when you want to describe the elements in a plot you can use something named "legend". in this case you can either add a 
corresponding label to each plot, to automatically be used when legend method is called or you can enter the labels within the
legend method itself. look at the example below:'''
x = np.linspace(0, 10, 500)
y1 = x
y2 = np.power(x, 2)
plt.plot(x, y1, label ='linear', c="g")
plt.plot(x, y2, label ='quadratic', c="r") 
plt.legend()
plt.show()
```

```python
'''or you can leave the label parameter blank and add the labels to the legend function itself.but be careful about the 
order of labels in the list.'''
x = np.linspace(0, 10, 500)
y1 = x
y2 = np.power(x, 2)
plt.plot(x, y1, c="g")
plt.plot(x, y2, c="r") 
plt.legend(['linear', 'quadratic'])
plt.show()
```

```python
'''legend will be placed automatically in the blankest location of the plot. but you can determine it's location too.'''
x = np.linspace(0, 10, 500)
y1 = np.sin(x)
y2 = np.power(x, 2)
plt.plot(x, y1, c="g")
plt.plot(x, y2, c="r") 
plt.legend(['sine', 'quadratic'], loc='lower right')
plt.show()
```
